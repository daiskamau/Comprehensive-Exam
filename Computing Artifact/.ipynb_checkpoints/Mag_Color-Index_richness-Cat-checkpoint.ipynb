```python
import glob,os
import numpy as np
import pandas as pd

import clmm
import GCRCatalogs as gcr
import astropy
from astropy.table import Table, setdiff, vstack

import matplotlib.pyplot as plt
```

```python
redmapper = gcr.load_catalog('cosmoDC2_v1.1.4_redmapper_v0.7.5')
```

```python
quantities = np.sort(redmapper.list_all_quantities())
print(len(quantities))
```

```python
# Cluster quantities
cl_quantities = [q for q in quantities if 'member' not in q]
print(cl_quantities, len(cl_quantities))
```

```python
# Member quantities
me_quantities = [q for q in quantities if 'member' in q]
print(me_quantities, len(me_quantities))
```

```python
# %%time
# Reading the richness table
zmin = 0.25 # 2378 if from 0
zmax = 0.26

cl_members0 = Table(redmapper.get_quantities(['cluster_id_member','ra_member','dec_member','mag_g_lsst_member', 'mag_i_lsst_member', 'mag_r_lsst_member', 'p_member'], filters=[ f'redshift_true_member > 0.2',f'redshift_true_member < 1', f'cluster_id_member==1']))
cl_members1 = Table(redmapper.get_quantities(['cluster_id_member','ra_member','dec_member','mag_z_lsst_member','mag_g_lsst_member', 'mag_i_lsst_member', 'mag_r_lsst_member', 'p_member'], filters=[ f'redshift_true_member > {zmin}',f'redshift_true_member < {zmax}', f'p_member>0.9']))
cl_members2 = Table(redmapper.get_quantities(['cluster_id_member','ra_member','dec_member','mag_z_lsst_member','mag_g_lsst_member', 'mag_i_lsst_member', 'mag_r_lsst_member', 'redshift_true_member','p_member'], filters=[ f'redshift_true_member > 0.2',f'redshift_true_member < 1', f'p_member>0.9']))
```

```python
plt.scatter(cl_members0['ra_member'], cl_members0['dec_member']) # cluster_id_member==1
```

```python
## Color-index <0.5 is blue(hot) ; >1.5 is red(cold) galaxy colour magnitude diagram
x = cl_members1['mag_i_lsst_member']
y0 = cl_members1['mag_g_lsst_member']-cl_members1['mag_r_lsst_member']
y1 = cl_members1['mag_r_lsst_member']-cl_members1['mag_i_lsst_member']

#find line of best fit
a0, b0 = np.polyfit(x, y0, 1)
print(a0,b0)
a1, b1 = np.polyfit(x, y1, 1)

fig, axs = plt.subplots(2, 1, sharex=True, figsize=(10,8))
# Remove horizontal space between axes
fig.subplots_adjust(hspace=0)

# Plot each graph, and manually set the y tick values
axs[0].scatter(x,y0, c='k', s=4)
axs[0].plot(x, a0*x+b0, color='red', linestyle='--', linewidth=2)
axs[0].set_yticks(np.arange(1, 1.5, 0.1)) 
axs[0].set_ylabel('g-r')

axs[1].scatter(x,y1, c='k', s=4)
axs[1].plot(x, a1*x+b1, color='red', linestyle='--', linewidth=2)
axs[1].set_yticks(np.arange(0.4, 0.8, 0.1)) 
axs[1].set_ylabel('r-i')

plt.show()
```
